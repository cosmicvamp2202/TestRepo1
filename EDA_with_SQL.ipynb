### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
import sqlite3
import pandas as pd

con = sqlite3.connect("my_data1.db")
query = "SELECT DISTINCT Launch_Site FROM SPACEXTABLE;"
pd.read_sql_query(query, con)
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'KSC'


```python
import pandas as pd
import sqlite3

con = sqlite3.connect("my_data1.db")
query = """
SELECT *
FROM SPACEXTABLE
WHERE Launch_Site LIKE 'KSC%'
LIMIT 5;
"""
pd.read_sql_query(query, con)
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
import sqlite3
import pandas as pd

# Connect to your database
con = sqlite3.connect("my_data1.db")

# SQL query
query = """
SELECT SUM(PAYLOAD_MASS__KG_) AS Total_Payload_Mass
FROM SPACEXTABLE
WHERE Customer = 'NASA (CRS)';
"""

# Execute and display result
df_result = pd.read_sql_query(query, con)
print(df_result)
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
import sqlite3
import pandas as pd

# Connect to the database
con = sqlite3.connect("my_data1.db")

# SQL query
query = """
SELECT AVG(PAYLOAD_MASS__KG_) AS Average_Payload_Mass
FROM SPACEXTABLE
WHERE Booster_Version = 'F9 v1.1';
"""

# Execute and display result
df_result = pd.read_sql_query(query, con)
print(df_result)
```

### Task 5

##### List the date where the succesful landing outcome in drone ship was acheived.


_Hint:Use min function_


```python
import sqlite3
import pandas as pd

# Connect to your database
con = sqlite3.connect("my_data1.db")

# SQL query to find the earliest successful drone ship landing
query = """
SELECT MIN(Date) AS First_Successful_Drone_Ship_Landing
FROM SPACEXTABLE
WHERE Landing_Outcome = 'Success (drone ship)';
"""

# Execute and display the result
df_result = pd.read_sql_query(query, con)
print(df_result)
```

### Task 6

##### List the names of the boosters which have success in ground pad  and have payload mass greater than 4000 but less than 6000


```python
import sqlite3
import pandas as pd

# Connect to the database
con = sqlite3.connect("my_data1.db")

# SQL query
query = """
SELECT Booster_Version, PAYLOAD_MASS__KG_, Landing_Outcome
FROM SPACEXTABLE
WHERE Landing_Outcome = 'Success (ground pad)'
  AND PAYLOAD_MASS__KG_ > 4000
  AND PAYLOAD_MASS__KG_ < 6000;
"""

# Execute and display the result
df_result = pd.read_sql_query(query, con)
print(df_result)
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
import sqlite3
import pandas as pd

# Connect to the database
con = sqlite3.connect("my_data1.db")

# SQL query to count total successful and failed missions
query = """
SELECT Mission_Outcome, COUNT(*) AS Total_Count
FROM SPACEXTABLE
GROUP BY Mission_Outcome;
"""

# Execute and display results
df_result = pd.read_sql_query(query, con)
print(df_result)
```

### Task 8



##### List all the booster_versions that have carried the maximum payload mass. Use a subquery.


```python

import sqlite3
import pandas as pd

# Connect to the database
con = sqlite3.connect("my_data1.db")

# SQL query using a subquery
query = """
SELECT Booster_Version, PAYLOAD_MASS__KG_
FROM SPACEXTABLE
WHERE PAYLOAD_MASS__KG_ = (
    SELECT MAX(PAYLOAD_MASS__KG_)
    FROM SPACEXTABLE
);
"""

# Execute and display the result
df_result = pd.read_sql_query(query, con)
print(df_result)
```

### Task 9


##### List the records which will display the month names, succesful landing_outcomes in ground pad ,booster versions, launch_site for the months in year 2017
**Note: SQLLite does not support monthnames. So you need to use substr(Date,6,2) for month, substr(Date,9,2) for date, substr(Date,0,5),='2017' for year.**


```python
import sqlite3
import pandas as pd
con = sqlite3.connect("my_data1.db")
query = """
SELECT
  CASE substr(Date,6,2)
    WHEN '01' THEN 'January'
    WHEN '02' THEN 'February'
    WHEN '03' THEN 'March'
    WHEN '04' THEN 'April'
    WHEN '05' THEN 'May'
    WHEN '06' THEN 'June'
    WHEN '07' THEN 'July'
    WHEN '08' THEN 'August'
    WHEN '09' THEN 'September'
    WHEN '10' THEN 'October'
    WHEN '11' THEN 'November'
    WHEN '12' THEN 'December'
  END AS Month_Name,
  substr(Date,9,2) AS Day,
  Date,
  Landing_Outcome,
  Booster_Version,
  Launch_Site
FROM SPACEXTABLE
WHERE substr(Date,0,5) = '2017'
  AND Landing_Outcome = 'Success (ground pad)'
ORDER BY substr(Date,6,2), substr(Date,9,2);
"""
df_result = pd.read_sql_query(query, con)
print(df_result)
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python

import sqlite3
import pandas as pd

# Connect to your database
con = sqlite3.connect("my_data1.db")

# SQL query
query = """
SELECT
  Landing_Outcome,
  COUNT(*) AS Outcome_Count
FROM SPACEXTABLE
WHERE Date BETWEEN '2010-06-04' AND '2017-03-20'
GROUP BY Landing_Outcome
ORDER BY Outcome_Count DESC;
"""

# Execute and display the result
df_result = pd.read_sql_query(query, con)
print(df_result)
```
